<font color='blue'> $$B = \{(1,2,4), (9,7,3), (2,c,8)\}$$

_Hint:_ here are many correct ways to do this, but here are two possible ways to find $c$:
1. Using a `for` loop in Python to test different values of $c$.
2. By using mathematical reasoning and comparing the first and third rows.

The value of c that ensures the vectors do not form a basis for $R^3$ is 4. Since all of the other values in that vector are just a linear multiple of the first vector by a factor of 2. Since vectors in a basis must be linearly independent, c=4 makes the vectors not fit this criteria.

<font color='blue'> &#9989;**Do this:** Complete the block of code to create a function `linearly_independent` that:
    \
(a) Gets the vectors as input in one list of lists.
    \
(b) Returns the span of the vectors and the dimension of the space they span.

````python
import sympy as sym
import numpy as np

def linearly_independent(X):
    V = sym.Matrix(X) #Make V in a sym Matrix
    V = sym.Matrix(V).rref()[0]        #Reduce to row echelon form
    V = np.array(V.tolist()) #Get it back to an np.array.
    V = V[~np.all(V == 0, axis=1)]   # Use np.all to delete the 0 rows.
    
    V=sym.Matrix(V)  # Make V a sym Matrix again and return it.
    print(f"The vectors span a space of dimension={V.shape[0]}")
    return V
````

## 2. Change of basis

## Review: Coordinate vectors
**Definition:** Given a basis $\mathcal{B}=\{\mathbf{v_1}, \mathbf{v_2}, ..., \mathbf{v_k}\}$ for a subspace $H \subseteq \mathbb{R}^n$, we define the coordinates of a vector $\mathbf{x} \in H$ relative to the basis $\mathcal{B}$ as the scalars $c_1, c_2, ..., c_k$ such that 
$$\mathbf{x}=c_1\mathbf{v_1}+ c_2 \mathbf{v_2} + ... + c_k\mathbf{v_k}.$$
So, the vector $[\mathbf{x}]_{\mathcal{B}} = \begin{bmatrix} c_1 \\ c_2 \\ ... \\ c_k\end{bmatrix}$ is referred to as the coordinate vector of $\mathbf{x}$ relative to $\mathcal{B}$.  

**Theorem:** Given $\mathcal{B}=\{\mathbf{v_1}, \mathbf{v_2}, ..., \mathbf{v_k}\}$ and $\mathcal{C}=\{\mathbf{w_1}, \mathbf{w_2}, ..., \mathbf{w_k}\}$ are two bases of the subspace $H$, then there is a unique $k \times k$ matrix $P_{\mathcal{C}\leftarrow\mathcal{B}}$ such that $$[\mathbf{x}]_{\mathcal{C}} = P_{\mathcal{C}\leftarrow\mathcal{B}}[\mathbf{x}]_{\mathcal{B}}. $$

There are several ways to determine this matrix $P_{\mathcal{C}\leftarrow\mathcal{B}}$, two of which include:
1. Gaussian elimination: $\left[\begin{matrix} \mathbf{w_1} & \mathbf{w_2} & ... & \mathbf{w_k}\end{matrix} \middle \vert \begin{matrix} \mathbf{v_1} & \mathbf{v_2} & ... & \mathbf{v_k}\end{matrix}\right] \sim_{RREF}  \left[\begin{matrix} I \end{matrix} \middle \vert P_{\mathcal{C}\leftarrow\mathcal{B}}\right]$.
   
3. $P_{\mathcal{C}\leftarrow\mathcal{B}} = P_{\mathcal{C}}^{-1}P_{\mathcal{B}}$, where $P_{\mathcal{B}} =\begin{bmatrix} \mathbf{v_1} & \mathbf{v_2} & ... & \mathbf{v_k}\end{bmatrix}$, and $P_{\mathcal{C}} =\begin{bmatrix} \mathbf{w_1} & \mathbf{w_2} & ... & \mathbf{w_k}\end{bmatrix}$.


<font color='blue'> &#9989;**Do this:** What is the coordinate vector of $u=\begin{bmatrix} 9\\-3\\21\end{bmatrix}$ relative to the basis $B=\left\{ \begin{bmatrix} 2\\0\\-1\end{bmatrix}, \begin{bmatrix} 0\\1\\3\end{bmatrix}, \begin{bmatrix} 1\\1\\1\end{bmatrix} \right\}$ in $\mathbb R^3$?

````python
# put your answer here

B = np.array([[2,0,1], [0,1,1],[-1,3,1]])
u = np.array([9,-3,21])

B_inv = np.linalg.inv(B)

ub = B_inv @ u

sym.Matrix(ub)
````

<font color='blue'> &#9989;**Do this:** Another basis in in $\mathbb R^3$ is the set $W=\left\{ \begin{bmatrix} 1\\1\\0\end{bmatrix}, \begin{bmatrix} 0\\1\\1\end{bmatrix}, \begin{bmatrix} 1\\0\\1\end{bmatrix} \right\}$. Find the transition matrix from $W$ to $B$.

````python
W = np.array([
    [1, 0, 1],
    [1, 1, 0],
    [0, 1, 1]
])

w1 = B_inv @ W[:,0]
w2 = B_inv @ W[:,1]
w3 = B_inv @ W[:,2]

Q = [w1, w2, w3]
Q = sym.Matrix(Q)

Q
````

<font color='blue'> &#9989;**Do this:**  Write a function in python such that:\
(a) The input is two bases of a vector space as list lists.\
(b) Return the transition matrix from the first basis to the second.

````python
# put your answer here
def transition_matrix(B1,B2):
    #Change the matrices to np.arrays
    B1=np.array(B1)
    B2=np.array(B2)
    # Concatenate them using the np.concatenate function
    B3 = np.concatenate((B1, B2), axis=1)
    #Turn it to a sym matrix and apply rref
    M= sym.Matrix(B3)
    M=M.rref()[0]
    #keep only the part of the matrix that you need.
    M=M[:,3:]
    return M
````
