## Lecture 1 Review

**Vector addition properties**

For any vectors $u$, $v$, and $w$ of the same size/dimension, we have the following properties:

* Vector addition is commutative: $u + v = v + u$.
* Vector addition is associative: $u + (v + w) = (u + v) + w$. We can therefore write both as $u + v + w$.
* Adding the zero vector to a vector has no effect: $u + 0 = 0 + u$. (The size of the zero vector must match the size of $u$).
* $u - u = 0$. Subtracting a vector from itself yields the zero vector.

**Scalar multiplication properties**

For any vectors $u$, $v$, and $w$, and scalars $a$ and $b$, we have the following properties:

* Scalar multiplication is commutative: $a\mathbf{u} = \mathbf{u}a$
* Scalar multiplication is associative: $(ab)\mathbf{u} = a(b\mathbf{u})$
* Scalar multiplication is distributive: $$a(\mathbf{u} + \mathbf{v}) = a\mathbf{u} + a\mathbf{v}$$ and $$(a+b)\mathbf{u}= a\mathbf{u} + b\mathbf{v}.$$

## Visualizing linear combinations

<font color='blue'> **Do this:** Below there is a piece of code that produces a graph of two vectors `u = [2, 1]` and `v = [1, 3]` along with a linear combination of them. Define two vectors and explore their linear combinations using the tool.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_lincomb(a, b): 
    #a and b are scalars
    # Define base vectors
    u = np.array([2, 1])
    v = np.array([1, 3])
    combo = a * u + b * v

    # Collect all vectors
    vectors = [u, v, combo]
    all_x = [vec[0] for vec in vectors]
    all_y = [vec[1] for vec in vectors]

    # Determine quadrant bounds
    x_min = min(0, min(all_x)) - 1
    x_max = max(0, max(all_x)) + 1
    y_min = min(0, min(all_y)) - 1
    y_max = max(0, max(all_y)) + 1

    # Set up plot
    plt.figure(figsize=(6, 6))
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)

    # Helper to draw arrows and annotate coordinates
    def draw_arrow(vec, color, label):
        plt.arrow(0, 0, vec[0], vec[1], head_width=0.2, head_length=0.3,
                  fc=color, ec=color, length_includes_head=True)
        coord_text = f"{label} = ({vec[0]:.1f}, {vec[1]:.1f})"
        plt.text(vec[0]*1.05, vec[1]*1.05, coord_text, color=color, fontsize=11)

    draw_arrow(u, 'blue', 'u')
    draw_arrow(v, 'green', 'v')
    draw_arrow(combo, 'red', f'{a}u + {b}v')

    # Truncate view to active quadrant
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.gca().set_aspect('equal')
    plt.grid(True)
    plt.title("Linear Combination in Active Quadrant")
    plt.show()
```

```python
#Example function call
plot_lincomb(1,2)
```
