<font color='blue'>9. By interpreting the RREF, what is the rank of $A$? How are the rank, dimension of the null space and the size of $A$ related?

The rank of A is equal to 2 (number of pivot rows in the rref). The rank, dimension of the null space, and size of A are related through the rank-nullity theorem, which says rank(A) + nullity(A) = number of columns of A. This is true here, because the rank is 2 and the nullity is also 2. This adds up to 4, which is the number of columns in A.

<font color='blue'>10. What is  $\dim\left(\text{null}(A^T)\right)$? Justify your answer.

The $dim(null(A^T))$ is equal to 1. Since the rank of the transposed matrix is going to be the same as the rank of the original matrix, from the rank-nullity theorem we can determine the $dim(null(A^T))$ to be equal to 1 from the equation $rank(A) + null(A) = N$ where $N$ is equal to the number of columns in A. In this case, we are inspecting $A^T$, so the number of columns would be 3.
