## 1.  Visualizing the problem
We can visualize the solution to a system of linear equations in a graph, as follows:

````python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
````

<font color=blue> **Question:** Is the following system consistent? If yes, how many solutions does it have?
    $$b+c=30$$
    $$25b+20c=690$$ </font>
    
<font color=blue> &#9989; To answer that, let's look at the graph of the system by completing the code in the cell below:

````python
c = np.linspace(0,20)

b1 = 30-c
b2 =  (690 - 20*c)/25 #complete this line and run the cell!
````

<font color=blue> Now run the cell below:

````python
plt.plot(c,b1)
plt.plot(c,b2)
plt.grid(True)
plt.xlabel('c ')
plt.ylabel('b', rotation=0)
````

**&#9989; Now answer the question by observing the graph.**

Yes, the graph is consistent (it only has one solution).

## 3.  Row reduction using python Sympy

The Python ``sympy`` library has a "reduced row echelon form" (rref) function that runs a much more efficient version of the Gauss elimination. To use the rref function you must first convert your matrix into a ``sympy.Matrix`` and then run the function. For example, let's do this for the following matrix $B$:

````python
import sympy as sym
import numpy as np

B =np.array([[ 50, 13, 30 ], [100, 26, 60 ],  [20.5, 25, 650]])
sym.Matrix(B).rref()[0]
````

**<font color=green> Remark:** Sometimes the above command would not give the correct RREF because of the numerical instability of the approximations. We can correct it, by extending the tolerance by higher powers of $10$ as follows. For some matrix $E$,
```
E = sym.Matrix(E)
rref_E = E.rref(iszerofunc=lambda x: abs(x) < 10**-10) #any value with absolute magnitude less than 10^{-10} is treated as zero.
rref_E[0]
```

<font color='blue'> &#9989; **Do this:** By row reducing the following augmented matrices, decide whether the systems have solutions or not. If they have solutions, how many are there? **Find the solutions by interpreting the RREF, if they exist**.

<font color='blue'> (a) $\left[\begin{matrix} 1 & -3& 7\\ -2& 1& -4 \\ 1& 2& 9\end{matrix} \middle\vert \begin{matrix} 0 \\ 0\\ 0\end{matrix} \right]$

````python
#put your answer here
A = np.array([[1, -3, 7, 0], [-2, 1, -4, 0], [1, 2, 9, 0]])
sym.Matrix(A).rref()[0]
````

The system has one solution, that being the zero vector. 

<font color='blue'> (b) $\left[\begin{matrix} 5 & 8 & 7\\ 0& 1& -1 \\ 1& 3& 0\end{matrix} \middle\vert \begin{matrix} 2 \\ -3\\ 2\end{matrix} \right]$

````python
# put your answer here
C = np.array([[5, 8, 7, 2], [0, 1, -1, -3], [1, 3, 0, 2]])
sym.Matrix(C).rref()[0]
````

The system has 0 solutions, as shown by the last row of the matrix. 0*x1 + 0*x2 + 0*x3 can never equal 1.

<font color='blue'> (c) $\left[\begin{matrix} 1 & 4& -5\\ 2& -1& 8 \end{matrix} \middle\vert \begin{matrix} 0 \\ 9\end{matrix} \right]$

````python
# put your answer here
D = np.array([[1, 4, -5, 0], [2, -1, 8, 9]])
sym.Matrix(D).rref()[0]
````

The system has infinitely many solutions, as there must be a free variable in order for it to be solved. 

<font color='blue'> (d) $\left[\begin{matrix} 1 & 2& 1\\ -3& -1& 2 \\ 0& 5& 3\end{matrix} \middle\vert \begin{matrix} 0 \\ 1\\ -1\end{matrix} \right]$

````python
# put your answer here
E = np.array([[1, 2, 1, 0], [-3, -1, 2, 1], [0, 5, 3, -1]])
sym.Matrix(E).rref()[0]
````

The system has one solution.

<font color='blue'> (e) $\left[\begin{matrix} 1 & 3& -3& 7\\ 0& 1& -4 &5\end{matrix} \middle\vert \begin{matrix} 0 \\ 0\end{matrix} \right]$

````python
# put your answer here
F = np.array([[1, 3, -3, 7, 0], [0, 1, -4, 5, 0]])
sym.Matrix(F).rref()[0]
````

The system has more variables than equations, so there must be a free variable(s) to solve. Because of this, there are infinitely many solutions.
