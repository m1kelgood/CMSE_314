## 1. Translating word problems as linear systems
#### The inconsiderate roommate:

Your roommate has used your credit card to make three types of purchases: online games, fast food, and taxi rides. Each type of purchase had a different cost, and now you are trying to figure out how much was spent on each.

This is what you know:

- Online Games: $\$15$ per game.
- Fast Food Orders: $\$10$ per order.
- Taxi Rides: $\$12$ per ride.
- Total Cost: $\$1200$.
 
Additionally, you know (somehow) there were 100 transactions in total, and there are 10 more fast food orders than online game purchases.

How many of each type of purchase (online games, fast food orders, and taxi ride) did your roommate make?

**&#9989; <font color=blue>Task:** <font color=blue>Create a linear system using augmented matrix, find its RREF, solve and interpret.

```python
#your answer here

import sympy as sym
import numpy as np

A= np.array([
    [15, 10, 12, 1200],
    [1, 1, 1, 100],
    [-1, 1, 0, 10]
]) #fill in the augmented matrix of the linear system


#compute the RREF of matrix A
sym.Matrix(A).rref()[0]
```

<font size=+3>&#9998;</font> Online games: 20 purchases, fast food: 30 purchases, taxi rides: 50 purchases.

## 2. Linear combinations

"_Is the vector $z$ a linear combination of the vectors $x$ and $y$?_"

This question is equivalent to "Does there exist constants $c_1$ and $c_2$ such that $z=c_1x+c_2y$?"

<font color=blue>**&#9989; Question** Is the vector $\begin{bmatrix}1 \\ -4\end{bmatrix}$ a linear combination of the vectors $\begin{bmatrix}-1 \\ 2\end{bmatrix}$ and $\begin{bmatrix}3 \\ -4\end{bmatrix}$? 

_Hint:_ set up a system and solve it.

```python
# your answer here
A = np.array([
    [-1, 3, 1],
    [2, -4, -4]
])

sym.Matrix(A).rref()[0]
```

$\begin{bmatrix}1 \\ -4\end{bmatrix}$ is a linear combination of the form -4 * $\begin{bmatrix}-1 \\ 2\end{bmatrix}$ - 1 * $\begin{bmatrix}3 \\ -4\end{bmatrix}$?

**&#9989;<font color=blue>Now, do this:** Run the following two cells. This will generate a widget with two sliders to select values for $c_1$ and $c_2$. The widget will plot: 
* the vector $c_1\vec{u}$ (a scalar multiple of $\vec{u}$) in blue with the "tail" at the origin
* the vector $c_2\vec{v}$ (a scalar multiple of $\vec{v}$) in red with the "tail" at the "tip" of the vector $c_1\vec{u}$,
* the target vector $\vec{w}$ in purple with the "tail" at the origin

By plotting the vectors $c_1\vec{u}$ and $c_2\vec{v}$ in the manner above (i.e. "tip to tail"), the sum of the two vectors, i.e. $c_1\vec{u}+c_2\vec{v}$ is the vector from the "tail" of the vector $c_1\vec{u}$ to the "tip" of the vector $y\vec{v}$. So if $c_1\vec{u}+c_2\vec{v} = \vec{w}$, the tip of the red vector will coincide with the tip of the purple vector.''

Set the values for the sliders for $c_1$ and $c_2$ such that $c_1\vec{u}+c_2\vec{v} = \vec{w}$, i.e. $c_1\begin{bmatrix}-1 \\ 2\end{bmatrix} + c_2\begin{bmatrix}3 \\ -4\end{bmatrix} = \begin{bmatrix}1 \\ -4\end{bmatrix}$.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Custom arrow plotting with scaled tip
def plot_vector(start, end, color):
    tip_width = 0.2
    tip_length = 0.2
    vec = end - start
    vec_length = np.linalg.norm(vec)
    if vec_length != 0:
        scaled_vec = vec * (vec_length - tip_length) / vec_length
    else:
        scaled_vec = np.zeros_like(vec)
        tip_width = 0
        tip_length = 0
    plt.arrow(start[0], start[1], scaled_vec[0], scaled_vec[1],
              head_width=tip_width, head_length=tip_length, fc=color, ec=color)

# Interactive column picture
def column_picture(u, v, target, step=0.5):
    min_x, max_x = -5.0, 5.0
    min_y, max_y = -5.0, 5.0

    def plot_linear_combination(a, b):
        plt.figure(figsize=(4, 4))
        plt.axis('equal')
        plt.grid(True)

        # Compute scaled vectors
        a_u = a * u
        b_v = b * v
        sum_vec = a_u + b_v

        # Set axis limits based on vector magnitudes
        bound_x = max(abs(a_u[0]), abs(b_v[0]), abs(target[0]), 5)
        bound_y = max(abs(a_u[1]), abs(b_v[1]), abs(target[1]), 5)
        plt.xlim(-bound_x - 1, bound_x + 1)
        plt.ylim(-bound_y - 1, bound_y + 1)

        # Plot vectors
        plot_vector(np.zeros(2), a_u, color='blue')         # a·u from origin
        plot_vector(a_u, a_u + b_v, color='red')            # b·v from tip of a·u
        plot_vector(np.zeros(2), target, color='purple')    # target from origin

        # Annotate tips
        plt.plot(*sum_vec, 'ro')  # Tip of b·v
        plt.plot(*target, 'mo')   # Tip of target

        # Title and labels
        plt.title(f'a·u + b·v = {sum_vec}\nTarget = {target}')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.axis('equal')
        plt.grid(True)
        plt.show()

    interact(plot_linear_combination,
             a=FloatSlider(value=1.0, min=min_x, max=max_x, step=step, description='a'),
             b=FloatSlider(value=1.0, min=min_y, max=max_y, step=step, description='b'))
```

```python
a = np.array([
    [1, 2, 3],
    [2, -1, 1]
])

sym.Matrix(a).rref()[0]
```

```python
# Example usage
u = np.array([1, 2])
v = np.array([2, -1])
target = np.array([3, 1])
column_picture(u, v, target)
```

## 3. Spanning vectors
"_Is a vector in a plane spanned by other vectors?_"

The previous example about linear combinations can be equally considered as follows. Given two vectors $\vec{u},\vec{v}$ that span a plane, is a third vector $\vec{w}$ on that plane? If $\vec{w}$ can be written as a linear combination of the other two the answer is yes, otherwise no. Let's see the following example.

<font color='blue'> &#9989; **Question**  Let $a_1=\begin{bmatrix} 1\\-2\\3 \end{bmatrix} $,  $a_2=\begin{bmatrix} 5\\-13\\-3 \end{bmatrix} $ and  $b=\begin{bmatrix} -3\\8\\1\end{bmatrix}$. Then span$\{a_1, a_2\}$ is a plane. Does $b$ belong in this plane?   

_Hint:_ Row reduce a linear system.

```python
w = np.array([
    [1, 5, -3],
    [-2, -13, 8],
    [3, -3, 1]
])

sym.Matrix(w).rref()[0]
```

<font size=+3>&#9998;</font> No, w cannot be written as a linear combination of the other two vectors.

## 4. Linear Independence

**Definition**: The vectors $v_1, v_2, \dots v_p$ are called **linearly independent** if the vector equation
$$
x_1v_1+x_2 v_2+ \dots+ x_p v_p=0
$$
has **only** the trivial solution, i.e. $x_1=x_2=\dots=x_p=0$.

These vectors are called **linearly dependent**, if there is at least one $x_j\neq 0$ such that $x_1v_1+x_2 v_2+ \dots+ x_p v_p=0$.

<font color='blue'> &#9989; **Question** Check whether the following vectors are linearly independent:
$\begin{bmatrix} 1\\2\\3 \end{bmatrix} $,  $\begin{bmatrix} 4\\5\\6 \end{bmatrix} $ and  $\begin{bmatrix} 2\\1\\0\end{bmatrix} $.

```python
#your code here
x = np.array([
    [1, 4, 2, 0],
    [2, 5, 1, 0],
    [3, 6, 0, 0]
])

sym.Matrix(x).rref()[0]
```

<font size=+3>&#9998;</font> The above vectors are linearly dependent.
