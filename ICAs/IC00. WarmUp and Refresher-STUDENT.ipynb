## 4. Numpy Arrays

NumPy is a fundamental Python library that provides support for large, multi-dimensional arrays and matrices. 
- Numpy includes a number of mathematical functions that are useful for working with these arrays.
- One of the most common uses of NumPy is to represent vectors in Python. 

A numpy **array** is a grid of values, all of the same type. An array is indexed by a tuple of nonnegative integers. 
The shape of an array is a tuple of integers giving the size of the array along each dimension.

- You can create a vector using the `array()` function from NumPy.


#### Creating Arrays

We can initialize numpy arrays from nested Python lists, and access elements using square brackets. Let's start with a one-dimensional array.

````python
import numpy as np
import sympy as sym

# 1-dimensional array
u = np.array([17, 24, 39])
````

````python
# check the type and shape. 
print(type(u))           
print(u.shape)
#or
print(np.shape(u))
````

````python
# access the the first two elements
print(u[0], u[1])
````

It is good practice to always check `.shape` before multiplying or solving systems. Now let's see a two-dimensional array.

````python
# 2-dimensional array
M = np.array([[1,2,3],[4,5,6]])
sym.Matrix(M)
````

Notice that the two layers of square brackets represent the nested structure necessary for creating a 2D array.

Like before, we can check the type and the shape of the array.

````python
# check the type and shape
print(type(M))
print(M.shape)

# extract certain elements
print(M[0, 0], M[0, 1], M[1, 0])

#Extract first row of matrix M
M[0,:]
````

#### Indexing/ slicing arrays

We can access particular elements of an array using the indexing system similar to lists. Let's try this with an image, which is a three-dimensional array. 

Carefully read and run the cells below.

````python
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np
import imageio.v2 as imageio

#from urllib.request import urlopen, urlretrieve
#from scipy.misc import imsave

url = 'https://res.cloudinary.com/miles-extranet-dev/image/upload/ar_16:9,c_fill,w_1000,g_face,q_50/Michigan/migration_photos/G21696/G21696-msubeaumonttower01.jpg'
im = imageio.imread(url) #3D Numpy array representing image

im[10,10,0] = 255 # Set the red channel value at pixel (10, 10) to maximum (255)
im[10,10,1] = 255 # Set the green channel value at pixel (10, 10) to maximum (255)
im[10,10,2] = 255 # Set the blue channel value at pixel (10, 10) to maximum (255)

#Show the image
plt.imshow(im);
````

````python
im[20,20,:] = 255 #sets all three color channels (Red, Green, Blue) of the pixel at row 20, column 20 to 255.

plt.imshow(im)
````

````python
cropped = im[0:50,0:50,:] #slice image to produce a 5050 pixel square from the top-left corner of the image
plt.imshow(cropped)
````

````python
cropped = im[50:,350:610,:] # Crop a specific region from the image
plt.imshow(cropped)
````

````python
red = im[:,:,0] # Extract the red channel from the image
plt.imshow(red)
plt.colorbar()
````

````python
#We isolate and visualize just the red channel
# We added the astype function to ensure that values are between 0-255
red_only = np.zeros(im.shape).astype(int) 
red_only[:,:,0] = red

plt.imshow(red_only)
````

````python
#We isolate and visualize just the green channel
green_only = np.zeros(im.shape).astype(int)
green_only[:,:,1] = im[:,:,1]

plt.imshow(green_only)
````

````python
#We isolate and visualize just the blue channel
blue_only = np.zeros(im.shape).astype(int)
blue_only[:,:,2] = im[:,:,2]

plt.imshow(blue_only)
````

#### Combining arrays/ Adding elements to arrays

- You can add elements to an array using the function `np.concatenate((array 1, array 2))`.
- You can also use the parameter `axis` to specify what sort of concatenation you are looking for. 

````python
array_1 = np.array([[1,2],[3,4]])
array_2 = np.array([[5,6]])
````

````python
np.concatenate((array_1,array_2), axis = 0)
````

````python
array_1
````

````python
# to turn array 2 from horizontal row to a vertical column, we use .T
# this is called transpose and we will formally study this later

np.concatenate((array_1,array_2.T), axis = 1)
````

## 6. The `random` package

This package is used to randomly generate numbers, when simulating data or initializing matrices. One could leverage it to test functions.

````python
import random

# random.random gives a value between 0 and 1
r1 = random.random()
print(r1)

# random.uniform gives a value between the two parameter bounds
r2 = random.uniform(4, 5)
print(r2)

# random.randint gives an integer between the two parameter bounds
r3 = random.randint(2,6)
print(r3)
````

One way to generate two lists or arrays in python, instead of writing them by hand, is the following. You can use the function to generate random lists and you can modify the function to get different results.
```
import random

def generate_random_lists(n):
    lists = []
    for i in range(n):
        v1 = [random.randint(-10, 10) for i in range(random.randint(1, 10))]
        v2 = [random.randint(-10, 10) for i in range(random.randint(1, 10))]
        lists.append((v1,v2))
    return lists

lists = generate_random_lists(5)
for v1,v2 in lists:
    print(f"v1={v1}, v2={v2}")
```
