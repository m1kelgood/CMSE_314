<font color='blue'> &#9989; **Do this:** Complete the following block of code to write a function that projects one vector onto another.

````python
import numpy as np
````

````python
def projection(v, u):
    u = np.array(u, dtype=float)
    v = np.array(v, dtype=float)
    if len(u) != len(v):
        raise ValueError("Vectors must have same dimension")

    y = np.dot(u, u)
    if y == 0:
        return np.zeros_like(u)   # or raise an error

    w = np.dot(v, u)
    return (w / y) * u
````

<font color='blue'> &#9989;**Do this:** We will create a function that implements the Gram-Schmidt process:\
This function takes a $m \times n$ Matrix $A$ with **linearly independent** columns as input and returns an $m \times n$ Matrix $G$ with orthogonal column vectors. 
    
<font color='blue'> The basic algorithm works as follows:

- <font color='blue'>`AT = transpose(A)` (this process works with the columns of the matrix so it is easier to work with the transpose. Think about a list of lists, it is easy to get a row (a list)).  
- <font color='blue'> Make a new empty list of the same size as ```AT``` and call it ```GT``` (G transpose)
- <font color='blue'>Loop index ```i``` over all of the rows in AT (i.e. columns of A) 

    - <font color='blue'>GT[i] is set to be  AT[i]
    - <font color='blue'>Loop index ```j``` from 0 to ```i```
        - GT[i] -= projection of AT[i] onto GT[j]
        
        
- <font color='blue'>```G = transpose(GT)```

<font color='blue'>Use the above definition as a template:

````python
def GramSchmidt(A):
    A = np.array(A, dtype=float)
    AT = A.T
    GT = np.zeros_like(AT)

    for i in range(AT.shape[0]):
        ui = AT[i].copy()

        for j in range(i):
            uj = GT[j]
            if np.dot(uj, uj) != 0:
                ui -= projection(ui, uj)

        GT[i] = ui

    return GT.T
````

<font color='blue'> &#9989;**Do this:** Create a new function called ```generalGS``` that checks first if the given vectors are linearly independent. If they are not, it keeps the linearly independent and returns an orthogonal basis for the space they span. 

````python
def generalGS(A):
    # Get the transpose of A
    A = np.array(A, dtype=float)
    AT = A.T

    # Use np.all() to delete the zero elements of AT (zero vectors)
    AT = AT[~np.all(AT == 0, axis=1)]
    
    # Set GT to be the first element of AT
    GT = AT[0].reshape(1, -1)
    
    # Keep a counter of the number of elements in GT. Currently we have 1.
    k = 1
    
    # Loop over the rows of AT
    for i in range(1, AT.shape[0]):
        proj = AT[i].copy()
        # Loop over the previous rows of GT to calculate the projection.
        for j in range(k):
            # Subtract the projection of AT[i] onto GT[j]
            proj -= projection(proj, GT[j])
        # If the element we got is not zero append it to the matrix.
        if not np.allclose(proj, np.zeros_like(proj)):   # to deal with approximations
            GT = np.append(GT, proj.reshape(1, -1), axis=0)
            k += 1
            
    # Transpose GT to get the orthogonal matrix G
    G = GT.T

    return G
````
