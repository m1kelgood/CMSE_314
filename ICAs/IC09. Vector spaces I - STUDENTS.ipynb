## 2. Basis

### Review
**Definition (basis):** If some vectors **SPAN** a space AND they are **LINEARLY INDEPENDENT**, then they form a **BASIS** for this space.

**Definition (rank):** The rank of a matrix $A$ is the number of leading ones in the reduced row echelon form of $A$. We denote it by $\text{rank}(A)$.
To compute this in Python, use the function `np.linalg.matrix_rank()`.

Now, We shall determine a basis for $\mathbb{R}^3$ and verify some equivalent statements about what a basis is.

<font color='blue'>&#9989;**Do this:** Create a $3 \times 3$ matrix $A$ where the columns of $A$ are basis vectors of $\mathbb{R}^3$. Use `numpy` functionality to perform the tasks 1-5 that follow.

```python
#your answer here
A = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

sym.Matrix(A)
```

<font color='blue'> 1. Calculate the determinant of matrix $A$ to verify that we have a basis. What will the determinant be if it's not a basis? Repeat this step until you have a basis.

```python
#your answer here
np.linalg.det(A)
```

Because the determinant is not equal to zero, the vectors form a basis. If it's not a basis, the determinant would be zero.

<font color='blue'> 2. Calculate the inverse of $A$ after you have found a basis. Why is the matrix invertible? (What is the meaning of this inverse matrix? To be answered in a few more classes.)

```python
#your answer here
np.linalg.inv(A)
```

The matrix is invertible because its determinant is nonzero.

<font color='blue'> 3. Calculate the reduced row echelon form of $A$. What do we expect to get, since we know the matrix is invertible?

```python
#your answer here
sym.Matrix(A).rref()[0]
```

We expect to get a rref where the rank is equal to the number of columns/rows, since we know the matrix is invertible (and therefore linearly independent).

<font color='blue'> 4. How many leading ones are there in the above reduced row echelon form? What did we expect since A is invertible? (What is the rank of A? According to the definition above?)


The rank is 3. There is one leading one in every column. We expected this, since A is invertible it means that the vectors are linearly independent.

<font color='blue'>5. Take the matrix $A$ that you created above and the vector $b=(1,3,2)$. What is the solution to $Ax=b$?  How many solutions do we expect this system to have?

```python
#your answer here
b = np.array([
    [1],
    [3],
    [2]
])

A_inv = np.linalg.inv(A)

X = A_inv @ b

sym.Matrix(X)
```

We expect this system to have exactly one solution, since the vectors are linearly independent.

### Summary
So, it turns out that a matrix whose column vectors form a basis has a lot of interesting properties, and therefore the following statements are equivalent.

0. The column vectors of $A$ form a basis for $\mathbb R^n$
1. $|A| \ne 0$ 
2. $A$ is invertible.
3. $A$ is row equivalent to $I_n$ (i.e. its reduced row echelon form is $I_n$)
4. $\text{rank}(A) = n$
5.  The system of equations $Ax = b$ has a unique solution.

Not all matrices follow the above statements, but the ones that do are used throughout linear algebra, so it is important to we know these properties.
