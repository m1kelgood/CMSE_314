## Question 3: Determinants (5 points)

 <font color=blue> Let $P$ be a $3\times3$ an invertible matrix and $A=\begin{bmatrix}0 & 1 & 0\\-k & 5 & -1 \\3k & 7 & 2 \end{bmatrix}$, for some real number $k$. Find 
    
(2 points) <font color=blue>   (a) the value(s) of $k$ for which the matrix $A$ is **not** invertible. [Hint: Calculate the determinant of $A$]
    
(3 points) <font color=blue> (b) If you set $k=1$, find the determinant of the determinant of the matrix $(P^{-1}AP)^{2026}$, where $P$ is some invertible matrix. [Hint: Do not look for a matrix $P$, but use the properties of the determinant instead]

#### (a)

```python
# Solution (done by hand, verifying with numpy)
a = np.array([
    [0, 1, 0],
    [0, 5, -1],
    [0, 7, 2]
])

np.linalg.det(a)
```

#### (b)

```python
# determinant of A when k = 1
a = np.array([
    [0, 1, 0],
    [-1, 5, -1],
    [3, 7, 2]
])

det = round(np.linalg.det(a), 2)
det
```

By the properties of the determinant, the $det(A)$ = $det(P^{-1}AP)$ [similarity rule], which is equal to -1, as shown above. An additional property of the determinant is the power rule, which states that $det(A)^x = (det(A))^x$, so if the determinant of A is equal to -1 and the power value is 2026, the resultant value of $det(P^{-1}AP) = 1$
