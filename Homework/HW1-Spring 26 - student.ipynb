**<font color=red>QUESTION 4d:</font>** <font color=blue> Given a line of the form $y=ax+b$ and a vector $\mathbf{u}=[c,d]$, complete the python function below that finds the projection of the vector onto the line. 

````python
import numpy as np
def proj_onto_line(a,b,c,d):
    u = np.array([c,d])
    w = np.array([a, a*a])
    if u.shape == w.shape:
        result = np.dot(u,w) / np.dot(w,w) * w
        return result
    else:
        print('The vectors are not of equal dimension')
````

**<font color=red>QUESTION 4e:</font>** <font color=blue> Test your function with the line $y=-4x+7$ and the vector $[3,-2]$

**<font color=red>QUESTION 5a:</font>**<font color= blue> Create a function called ```are_orthogonal``` that takes two vectors as arguments and checks if two vectors are orthogonal and prints a message informing us. Make sure the function works for vectors of the same size.

````python
def are_orthogonal(u,v):
    u=np.array(u)
    v=np.array(v)
    if u.shape == v.shape:
        if np.dot(u,v) == 0:
            print('The vectors are orthogonal')
        else:
            print('The vectors are NOT orthogonal')
    else:
        print('Not of same size')
````

**<font color=red>QUESTION 5b:</font>**<font color= blue>  Create a function called ```my_proj``` that takes two vectors as arguments and calculates the projection of one of the vectors onto the other.

````python
def my_proj(u,v):
    u = np.array(u)
    v = np.array(v)
    if u.shape == v.shape:
        result = np.dot(u,v) / np.dot(v,v) * v
        return result
    else:
        print('The vectors are not of equal dimension')
````
