```python
import numpy as np
import sympy as sym
```

```python
P = np.array([
    [0.45, 0.65],
    [0.55, 0.35]
]) 

#        1       2
# P = 1 [0.45  0.65]
#     2 [0.55  0.35]
#input state transition matrix as P
sym.Matrix(P)

##### QUESTION 1#####
# Type current state vector x provided:
x=np.matrix([[0.55], [0.45]]) 
sym.Matrix(x)

# Probability distribution of fans after 1 season:
x2 = P @ x
sym.Matrix(x2)
```

```python
##### QUESTION 2 #####

# Compute eigenvalues and eigenvectors of the state transition matrix
eigvals,eigvecs = np.linalg.eig(P) 
eigvals, sym.Matrix(eigvecs)
```

```python
# fill in the index of the eigenvector of your matrix associated with an eigenvalue of 1.
# store this as a new vector called steadystate
steadystate=eigvecs[:, 0]
steadystate
```

```python
# Normalize your steady state vector:
steadystate = steadystate / sum(steadystate)

# Now insert the entire population size to find the long-run number of Michigan State fans:
steadystate = steadystate * 10e6
steadystate[0]
```
