**Task** <br>
Construct a 2×2 transformation matrix, T, that first rotates a point 45 degrees counterclockwise and then performs a shear of factor 1.5 along the y-direction. Apply the matrix T to the image provided and make a plot of the transformed image.

Which of the following is closest to the new coordinates of the top right-hand corner of the original image? 

```python
## NOW ANSWER THE QUESTION IN A NEW CELL:
h, w = banner.shape[:2] # Get the image height (h) and width (w)

theta = 45 # input rotation angle in degrees
t= np.radians(theta) # Convert degrees into radians for trigonometric functions
sin = np.sin(t) #compute sine of angle t
cos = np.cos(t) #compute cosine of angle t

# Create the 2×2 rotation matrix
R = np.array([
    [cos, -sin],
    [sin, cos]
])

# Create the shear matrix
S = np.array([
    [1, 0],
    [1.5, 1]
])

# Combine transformations: rotate first, then shear
T = S @ R


# Run the rest of the commands below exactly as they are:

coords = np.array([[x, y] for y in range(h) for x in range(w)]).T # Create coordinate matrix using all coordinate pairs (x,y) in the image

new_coords = T @ coords # Apply the transformation matrix to all pixel coordinates

# Plot the transformed pixels with their original colors
plt.scatter(new_coords[0], new_coords[1], c=banner.reshape(-1,3)/255, s=1)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
# plt.savefig("transformed.png")
plt.show()
```
